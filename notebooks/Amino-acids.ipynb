```python
import numpy as np
import matplotlib.pyplot as plt
from pepdata import *

RED = "#EB6231"
YELLOW = "#E29D26"
BLUE = "#5D80B4"
LIGHTGREEN = "#6ABD9B"
GREEN = "#8FB03E"

nucleotides = "ACGT"
codontable = {
    'ATA': 'I', 'ATC': 'I', 'ATT': 'I', 'ATG': 'M',
    'ACA': 'T', 'ACC': 'T', 'ACG': 'T', 'ACT': 'T',
    'AAC': 'N', 'AAT': 'N', 'AAA': 'K', 'AAG': 'K',
    'AGC': 'S', 'AGT': 'S', 'AGA': 'R', 'AGG': 'R',
    'CTA': 'L', 'CTC': 'L', 'CTG': 'L', 'CTT': 'L',
    'CCA': 'P', 'CCC': 'P', 'CCG': 'P', 'CCT': 'P',
    'CAC': 'H', 'CAT': 'H', 'CAA': 'Q', 'CAG': 'Q',
    'CGA': 'R', 'CGC': 'R', 'CGG': 'R', 'CGT': 'R',
    'GTA': 'V', 'GTC': 'V', 'GTG': 'V', 'GTT': 'V',
    'GCA': 'A', 'GCC': 'A', 'GCG': 'A', 'GCT': 'A',
    'GAC': 'D', 'GAT': 'D', 'GAA': 'E', 'GAG': 'E',
    'GGA': 'G', 'GGC': 'G', 'GGG': 'G', 'GGT': 'G',
    'TCA': 'S', 'TCC': 'S', 'TCG': 'S', 'TCT': 'S',
    'TTC': 'F', 'TTT': 'F', 'TTA': 'L', 'TTG': 'L',
    'TAC': 'Y', 'TAT': 'Y', 'TAA': 'X', 'TAG': 'X',
    'TGC': 'C', 'TGT': 'C', 'TGA': 'X', 'TGG': 'W'}
assert len(codontable.keys()) == 64, "There is not 64 codons in the codon table"

codons = list(codontable.keys())
assert len(codons) == 64, "There is not 3 stop codons in the codon table"
assert not [n for n in "".join(codons) if (n not in nucleotides)], "There is a codon with an unrecognized nucleotide"

amino_acids_set = set(codontable.values())
amino_acids_set.remove('X')

nbr_weak = np.array([c.count("A") + c.count("T") for c in codons])
nbr_strong = 3. - nbr_weak
```

## Plot of $\lambda_{obs}$ and %AT for all amino-acids

```python
points = 50
at_pct = np.zeros(points)
gc_pct = np.zeros(points)
mut_bias_obs = np.zeros(points)

for aa in sorted(amino_acids_set):
    name = amino_acid.index_to_long_name(amino_acid.letter_to_index(aa))

    mut_bias_range = np.logspace(-1, 1, points)
    for mut_bias_id, mut_bias in enumerate(mut_bias_range):
        codon_frequencies = np.power(mut_bias, nbr_weak)

        fitness_codons = np.array([codontable[codon] == aa for codon in codons])
        codon_frequencies *= fitness_codons
        codon_frequencies /= np.sum(codon_frequencies)
        
        at_pct[mut_bias_id] = np.sum(codon_frequencies * nbr_weak) / 3
        gc_pct[mut_bias_id] = np.sum(codon_frequencies * nbr_strong) / 3
        mut_bias_obs[mut_bias_id] = at_pct[mut_bias_id] / gc_pct[mut_bias_id]

    my_dpi = 128
    fig = plt.figure(figsize=(960 / my_dpi, 720 / my_dpi), dpi=my_dpi)
    plt.subplot(211)
    plt.plot(mut_bias_range, mut_bias_obs, '--', label='$\lambda_{obs}$', color=RED)
    plt.plot(mut_bias_range, mut_bias_range, alpha=0.75, label='$\lambda$', color=BLUE)
    plt.xscale('log')
    plt.yscale('log')
    plt.ylabel('$\lambda_{obs}$')
    plt.title(name)
    plt.legend()
    plt.tight_layout()
    plt.subplot(212)
    plt.plot(mut_bias_range, np.ones(points) / 3, '--', alpha=0.5, linewidth=1, color='grey')
    plt.plot(mut_bias_range, 2 * np.ones(points) / 3, '--', alpha=0.5, linewidth=1, color='grey')
    plt.plot(mut_bias_range, at_pct, '--', label='%AT$_{obs}$', color=RED)
    plt.plot(mut_bias_range, mut_bias_range / (mut_bias_range + 1), alpha=0.75, label='%AT', color=BLUE)
    plt.xscale('log')
    plt.ylim((0,1))
    plt.xlabel('$\lambda$')
    plt.ylabel('%AT')
    plt.legend()
    plt.tight_layout()
    plt.show()

    print("{0} ({1})".format(name, aa))
    aa_codons = [c for c in codons if codontable[c] == aa]
    print("{0} codons : ".format(len(aa_codons)) + ", ".join(aa_codons))
    print("l=0.1: ".format(len(aa_codons)) + ", ".join(aa_codons))
    print("hydropathy: {0}".format(amino_acid.hydropathy(aa)))
    print("volume: {0}".format(amino_acid.volume(aa)))
    print("polarity: {0}".format(amino_acid.polarity(aa)))
    print("pK_side_chain: {0}".format(amino_acid.pK_side_chain(aa)))
    print("prct_exposed_residues: {0}".format(amino_acid.prct_exposed_residues(aa)))
    print("hydrophilicity: {0}".format(amino_acid.hydrophilicity(aa)))
    print("accessible_surface_area: {0}".format(amino_acid.accessible_surface_area(aa)))
    print("local_flexibility: {0}".format(amino_acid.local_flexibility(aa)))
    print("accessible_surface_area_folded: {0}".format(amino_acid.accessible_surface_area_folded(aa)))
    print("refractivity: {0}".format(amino_acid.refractivity(aa)))
```

## Correlation between %AT and severals statistics

```python
import statsmodels.api as sm

aa_list = sorted(amino_acids_set)
for mut_bias in [0.1, 1.0, 10.0]:

    aa_at_pct = []
    
    for aa in sorted(amino_acids_set):
        codon_frequencies = np.power(mut_bias, nbr_weak)
        codon_frequencies *= np.array([codontable[codon] == aa for codon in codons])
        codon_frequencies /= np.sum(codon_frequencies)
        
        aa_at_pct.append(np.sum(codon_frequencies * nbr_weak) / 3)

    aa_stat = [amino_acid.hydrophilicity(aa) for aa in aa_list]

    my_dpi = 128
    fig = plt.figure(figsize=(960 / my_dpi, 480 / my_dpi), dpi=my_dpi)
    plt.scatter(aa_at_pct, aa_stat)
    idf = np.linspace(min((min(aa_at_pct), 0)), max(aa_at_pct), 30)
    model = sm.OLS(aa_stat, sm.add_constant(aa_at_pct))
    results = model.fit()
    b, a = results.params[0:2]
    plt.plot(idf, a * idf + b, 'r-', label=r"$y={0:.3g}x {3} {1:.3g}$ ($r^2={2:.3g})$".format(
                 float(a), abs(float(b)), results.rsquared, "+" if float(b) > 0 else "-"), color=BLUE)
    plt.xlim((0,max(aa_at_pct)))
    plt.ylabel("hydrophilicity")
    plt.xlabel('%AT$_{obs}$')
    plt.legend()
    plt.title('$\lambda={0}$'.format(mut_bias))
    plt.tight_layout()
    plt.show()
```

## Plot $\lambda_{obs}$ and %AT for a mixture

```python
points = 50
at_pct = np.zeros(points)
gc_pct = np.zeros(points)
mut_bias_obs = np.zeros(points)

aa_list = ["A", "K"]
aa_names = [amino_acid.index_to_long_name(amino_acid.letter_to_index(aa)) for aa in aa_list]
mut_bias_range = np.logspace(-1, 1, points)
for mut_bias_id, mut_bias in enumerate(mut_bias_range):
    for aa in aa_list:        
        codon_frequencies = np.power(mut_bias, nbr_weak)
        fitness_codons = np.array([codontable[codon] == aa for codon in codons])
        codon_frequencies *= fitness_codons
        codon_frequencies /= np.sum(codon_frequencies)

        at_pct[mut_bias_id] += np.sum(codon_frequencies * nbr_weak) / (3 * len(aa_list))
        gc_pct[mut_bias_id] += np.sum(codon_frequencies * nbr_strong) / (3 * len(aa_list))
    mut_bias_obs[mut_bias_id] = at_pct[mut_bias_id] / gc_pct[mut_bias_id]

my_dpi = 128
fig = plt.figure(figsize=(960 / my_dpi, 720 / my_dpi), dpi=my_dpi)
plt.subplot(211)
plt.plot(mut_bias_range, mut_bias_obs, '--', label='$\lambda_{obs}$', color=RED)
plt.plot(mut_bias_range, mut_bias_range, alpha=0.75, label='$\lambda$', color=BLUE)
plt.xscale('log')
plt.yscale('log')
plt.ylabel('$\lambda_{obs}$')
plt.title(", ".join(aa_names))
plt.legend()
plt.tight_layout()
plt.subplot(212)
plt.plot(mut_bias_range, np.ones(points) / 3, '--', alpha=0.5, linewidth=1, color='grey')
plt.plot(mut_bias_range, 2 * np.ones(points) / 3, '--', alpha=0.5, linewidth=1, color='grey')
plt.plot(mut_bias_range, at_pct, '--', label='%AT$_{obs}$', color=RED)
plt.plot(mut_bias_range, mut_bias_range / (mut_bias_range + 1), alpha=0.75, label='%AT', color=BLUE)
plt.xscale('log')
plt.ylim((0,1))
plt.xlabel('$\lambda$')
plt.ylabel('%AT')
plt.legend()
plt.tight_layout()
plt.show()

for aa, name in zip(aa_list, aa_names):
    print(name)
    aa_codons = [c for c in codons if codontable[c] == aa]
    print("{0} codons : ".format(len(aa_codons)) + ", ".join(aa_codons))
```

## Plot $\lambda_{obs}$ and %AT for a protein

```python
import os

aa_char_to_int = {v: k for k, v in enumerate(sorted(amino_acids_set))}
aa_char_to_int['X'] = 20
amino_acids = "".join(amino_acids_set) + "X"
assert len(amino_acids) == 21, "There is not 21 amino-acids in the codon table"

aa_table = {}
codon_to_aa = [None] * len(codons)
for index, codon in enumerate(codons):
    aa = codontable[codon]
    if aa not in aa_table:
        aa_table[aa] = []
    aa_table[aa].append(index)
    codon_to_aa[index] = aa_char_to_int[aa]
assert len(aa_table) == 21, "There is not 21 amino-acids in the aa table"

points = 100
for cds in ['random', 'gal4', 'lactamase', 'np', 'ha']:
    preferences_list = []

    file_path = '/home/thibault/projet_codon/preferences/{0}.txt'.format(cds)
    if os.path.isfile(file_path):
        with open(file_path, 'r') as prefs_file:
            header = prefs_file.readline()
            for line in prefs_file:
                prefs = [float(f) for f in line.replace('\n', '').replace('\t', ' ').split(' ')[3:]]
                prefs.append(0.0)
                preferences_list.append(np.array(prefs)/np.sum(prefs))
    else:
        for i in range(200):
            sigma_2 = np.random.uniform(0, 4)
            assert 0 <= sigma_2 <= 4
            sigma = np.sqrt(sigma_2)
            pref_aa = np.exp(np.random.normal(loc=0, scale=sigma, size=len(amino_acids)))
            pref_aa[-1] = 0.0
            pref_aa /= np.sum(pref_aa)
            preferences_list.append(pref_aa)
    
    mut_bias_range = np.logspace(-3, 3, points)
    population_sizes = np.logspace(0, 2, 5)
    mut_bias_pop = {}
    
    for population_size in population_sizes:
        
        nbr_sites = len(preferences_list)
        at_pct = np.zeros((points, nbr_sites))
        gc_pct = np.zeros((points, nbr_sites))

        for mut_bias_id, mut_bias in enumerate(mut_bias_range):
            
            for site, preferences in enumerate(preferences_list):
                codon_frequencies = np.power(mut_bias, nbr_weak)
                pref_codons = [preferences[codon_to_aa[i]] for i in range(len(codons))]
                codon_frequencies *= np.power(pref_codons, population_size)
                codon_frequencies /= np.sum(codon_frequencies)
                
                at_pct[mut_bias_id, site] = np.sum(codon_frequencies * nbr_weak)
                gc_pct[mut_bias_id, site] = np.sum(codon_frequencies * nbr_strong)

        mut_bias_obs = np.sum(at_pct, axis=1) / np.sum(gc_pct, axis=1)
        mut_bias_pop[population_size] = mut_bias_obs

    my_dpi = 128
    fig = plt.figure(figsize=(960 / my_dpi, 720 / my_dpi), dpi=my_dpi)
    plt.subplot(211)
    plt.plot(mut_bias_range, mut_bias_range, alpha=0.75, label='$\lambda$', color=BLUE)
    for pop in population_sizes:
        plt.plot(mut_bias_range, mut_bias_pop[pop], '--', label='$N_e$={0}'.format(pop))
    plt.xscale('log')
    plt.yscale('log')
    plt.ylabel('$\lambda_{obs}$')
    plt.title(cds)
    plt.legend()
    plt.tight_layout()
    plt.subplot(212)
    plt.plot(mut_bias_range[:-1], np.ones(points - 1), label='1', color='black')
    for pop in population_sizes:
        d_mut_bias_obs = np.log(mut_bias_pop[pop][1:]) - np.log(mut_bias_pop[pop][:-1])
        d_mut_bias = np.log(mut_bias_range[1:]) - np.log(mut_bias_range[:-1])
        plt.plot(mut_bias_range[:-1], d_mut_bias_obs / d_mut_bias, '--', label='$N_e$={0}'.format(pop))
    plt.xscale('log')
    plt.ylabel('d$ln(\lambda_{obs})$/d$ln(\lambda)$')
    plt.legend()
    plt.tight_layout()
    plt.show()
```
